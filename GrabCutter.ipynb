```python
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas as pd
from skimage.io import imread, imsave
from skimage import img_as_ubyte
from skimage.transform import resize
import skimage
import cv2
from grabcutter import GrabCutter
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#sample use
im_url='runner.jpg'
im_size=imread(im_url).shape
box= '100,0,300,350'
#grabcut with bounding box
original_image, segmented=GrabCutter()(im_url, box)
```

```python
#visualize initial grabcut result
fig, ax=plt.subplots(1,2, figsize=(10,10))
ax[0].imshow(original_image.get()[:,:,::-1])
ax[1].imshow(segmented[:,:,::-1])
```

```python
#grabcut with foreground/background iteration
"""
keys:
f: to select foreground
b: to select background
r: to rerun grabcut
spacebar: to terminate grabcut
+/-: to increase/decrease brush size
"""
_, segmented=GrabCutter()(im_url, '', with_refine=True)
```

```python
#to visualize final grabcut result
plt.imshow(segmented[:,:,::-1])
```
